```python
import pandas as pd
import json
import requests
import numpy as np
```

```python
response = requests.get("https://raw.githubusercontent.com/Papagoat/brain-assessment/main/restaurant_data.json")
if response.ok:
    json_data = response.json()

else: 
    print("No response detected")
```

```python
test_json_df = pd.json_normalize(json_data)
test_json_df.head()
```

```python
test_json_df["restaurants"][0]
```

```python
restaurant_df = pd.json_normalize(json_data, 'restaurants')
restaurant_df.head()
```

```python
len(restaurant_df)
```

```python
country_code_df = pd.read_excel("Country-Code.xlsx")
country_code_df
```

```python
merged_df = pd.merge(
    restaurant_df, country_code_df, 
    how='left', left_on='restaurant.location.country_id', 
    right_on='Country Code'
)
merged_df.head()
```
