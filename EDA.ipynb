# Question 2

```python
# Examine the structure of the values within the column "restaurant.zomato_events"
def check_length_list(value):
    if value is np.nan:
        return np.nan
    elif len(value) > 1:
        return value
    else:
        return np.nan

multiple_event_df = combined_df["restaurant.zomato_events"].apply(check_length_list).dropna()
multiple_event_df.iloc[0]
```

```python
# Expand event column vertically
explode_events_restaurant_df = combined_df.explode("restaurant.zomato_events")

print(len(explode_events_restaurant_df))
```

```python
# Unpack the dictionaries 
expnd_events_restaurant_df = pd.json_normalize(
    explode_events_restaurant_df['restaurant.zomato_events']
)
print(len(expnd_events_restaurant_df))
```

```python
merged_events_df = expnd_events_restaurant_df.join(explode_events_restaurant_df)
merged_events_df.head()
```

```python
merged_events_df = merged_events_df.rename(columns={"restaurant.zomato_events": "zomato_events"})
```

```python

merged_events_df['event.start_date'] = pd.to_datetime(merged_events_df['event.start_date'])
merged_events_df['event.end_date'] = pd.to_datetime(merged_events_df['event.end_date'])
```

```python
condition_startdate = (merged_events_df['event.start_date'].dt.year < 2019) | \
    ((merged_events_df['event.start_date'].dt.year == 2019) & (merged_events_df['event.start_date'].dt.month <= 4))


condition_enddate = (merged_events_df['event.end_date'].dt.year > 2019) | \
    ((merged_events_df['event.end_date'].dt.year == 2019) & (merged_events_df['event.end_date'].dt.month >= 4))

# Apply conditions
filtered_event_df = merged_events_df[condition_enddate & condition_startdate]
filtered_event_df.head()
```

```python
print(len(filtered_event_df))
```

```python
examine_photo_df = filtered_event_df["event.photos"].apply(check_length_list).dropna()
examine_photo_df.iloc[0]
```

```python
# Expand event column vertically
explode_photos_df = filtered_event_df.explode("event.photos")

len(explode_photos_df)
```

```python
explode_photos_df.head()
```

```python

# Unpack the dictionaries 
expnd_photos_restaurant_df = pd.json_normalize(
    explode_photos_df['event.photos']
)

len(expnd_photos_restaurant_df)
```

```python
merged_photos_df = explode_photos_df.join(expnd_photos_restaurant_df)
len(merged_photos_df)
```

```python
merged_photos_df.head()
```

```python
final_columns = [
    "event.event_id", "Restaurant Id", "Restaurant Name", 
    "restaurant.photos_url", "event.title", "event.start_date", "event.end_date"
]
final_events_df = merged_photos_df[final_columns]

columns_to_rename = {
    "event.event_id": "Event Id", 
    "restaurant.photos_url": "Photo URL",
    "event.title": "Event Title", 
    "event.start_date": "Event Start Date",
    "event.end_date": "Event End Date"
}

# Rename columns for readability
question2_df = final_events_df.rename(columns=columns_to_rename)

question2_df.head()
```

```python
filtered_df = restaurant_df[
    restaurant_df['restaurant.user_rating.rating_text'].isin(
        ['Excellent', 'Very Good', 'Good', 'Average', 'Poor']
    )
]
filtered_df.head()
```

```python
filtered_df['restaurant.user_rating.aggregate_rating'] = filtered_df['restaurant.user_rating.aggregate_rating'].apply(pd.to_numeric)
filtered_df = filtered_df.rename(columns={"restaurant.user_rating.rating_text":"Rating Text"})
```

```python
rating_threshold = filtered_df.groupby(
    'Rating Text', as_index=False)['restaurant.user_rating.aggregate_rating'].agg(['min', 'max'])

rating_threshold
```
